```
from __future__ import print_function, division
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import torchvision
from torchvision import datasets, models, transforms
```

```
model_list = []
for model in dir(models):
    if model[0].islower():
        if hasattr(getattr(models, model), '__call__'):
            print(f"Loading Pretrained Model: {model}")
            try:
                model_list.append((model, getattr(models, model)(pretrained=True)))
            except:
                print("Download failed!")
                print(f"No checkpoint is available for model: {model}\n")
                
```

```
len(model_list)
```

```
for model_name, model in model_list:
    print("=================================== " + model_name + " ===================================")
    print(model, "\n\n")
```

```
dont_have_classifier = []
have_classifier = []

print("Models containing classifier attribute\n")
for model_name, model in model_list:
    try:
        if getattr(model, "classifier"):
            print(model_name)
            have_classifier.append((model_name, model))
    except:
        dont_have_classifier.append((model_name, model))
```

```
print("Models not containing classifier attribute\n")
print(*list(list(zip(*dont_have_classifier))[0]), sep="\n")
```

```
for model_name, model in have_classifier:
    print(model_name)
    print(model.classifier)
    print()
```

```
for model_name, model in dont_have_classifier:
    print(model_name)
    print(model.fc)
    print()
```

```
from torchsummary import summary

for model_name, model in model_list:
    try:
        print("============================ " + model_name + " ============================")
        print(summary(model, (3, 224, 224), 32), "\n\n")
    except:
        print("torchsummary doesn't support DenseLayer. See: https://github.com/sksq96/pytorch-summary/pull/116 \n\n")
        pass
```
